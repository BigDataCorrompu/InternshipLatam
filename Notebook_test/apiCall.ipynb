```python
import requests
import os
from dotenv import load_dotenv
import base64

load_dotenv('../.env')

api_key    = os.getenv('CAREERJET_APP_KEY')
user_ip    = os.getenv('SERVER_IP')
user_agent = os.getenv('CAREERJET_USER_AGENT', 'InternshipLatam/1.0')

credentials = base64.b64encode(f"{api_key}:".encode()).decode()

response = requests.get(
    "https://search.api.careerjet.net/v4/query",
    headers={
        "Authorization": f"Basic {credentials}",
        "Referer": os.getenv('CAREERJET_REFERER', '')
    },
    params={
        "keywords": "data engineer",
        "location": "Santiago",
        "user_ip": user_ip,
        "user_agent": user_agent,
        "pagesize": 10
    }
)

print(f"Status : {response.status_code}")
print(f"Jobs   : {len(response.json().get('jobs', []))}")
print(response.json().keys())
```

```python
import sys
sys.path.append('../ingestion/python/src')

from mapper import JsearchMapper, JobMapper
from database import Database
from utils import load_json
from dataclasses import astuple
import os
from dotenv import load_dotenv

load_dotenv('../.env')

# Charge le fichier du 2026-06-09
blocks = load_json('../ingestion/raw/2026-06-09/api_jsearch', 'jsearch_search')

mapper   = JsearchMapper()
all_data = []

for block in blocks:
    params = block.get("params", {})
    data   = block.get("data", [])
    for d in data:
        row = mapper.normalise(data=d, metaData=params)
        all_data.append(row)

result = mapper.structureToBulkInsert(all_data)

db = Database(
    db_host     = os.getenv('DB_HOST'),
    db_name     = os.getenv('DB_NAME'),
    db_user     = os.getenv('DB_USER'),
    db_password = os.getenv('DB_PASSWORD'),
    db_sslmode  = os.getenv('DB_SSLMODE', 'require'),
    db_channelbinding = os.getenv('DB_CHANNELBINDING', 'disable'),
)

db.bulk_insert(
    'raw.job_offer',
    result['columns'],
    result['data'],
    onConflict='update',
    conflict_columns=['id_job']
)

print(f"Inséré : {len(all_data)} offres")
```
