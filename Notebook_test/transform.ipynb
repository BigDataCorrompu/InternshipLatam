```python
import json
import pandas as pd

RUN_DIR = "../ingestion/python/results/005/"

# Chemin d'extraction direct par fichier
EXTRACT = {
    "careerjet_search":        lambda r: r["raw"]["jobs"],
    "jsearch_search":          lambda r: r["raw"]["data"]["jobs"],
    "jsearch_company_salary":  lambda r: r["raw"]["data"],
    "jsearch_salary":          lambda r: r["raw"]["data"],
    "jsearch_details":         lambda r: r["raw"]["data"],
}

def load(name: str) -> pd.DataFrame:
    with open(RUN_DIR + name + ".json", encoding="utf-8") as f:
        raw = json.load(f)
    return pd.DataFrame(EXTRACT[name](raw))


```
