```python
from langdetect import detect, DetectorFactory

# Force le résultat à être stable et reproductible
DetectorFactory.seed = 0

donnees_brutes = ["espanol", "es", "spanish", "python", "C", "C#", "français", "en", "UNKNOWN", "C#", "python", "portuguese"]

def normaliser_langue_detect(texte):
    # Nettoyage de base
    txt = str(texte).strip().lower()
    
    # Écarter immédiatement les intrus informatiques et les codes trop courts
    intrus = {"python", "c#", "c", "unknown"}
    if txt in intrus or len(txt) < 2:
        return None
        
    try:
        # detect() lit le mot ("français", "spanish", "espanol") et retourne son code ISO 639-1
        return detect(txt)
    except Exception:
        return None

# Application sur la liste
resultat_clean = [normaliser_langue_detect(item) for item in donnees_brutes]

# Suppression des valeurs None et des doublons
resultat_final = list(set(code for code in resultat_clean if code is not None))

print(resultat_final)
# Sortie : ['fr', 'pt', 'es', 'en']
```

```python
import locale

donnees_brutes = ["espanol", "es", "spanish", "python", "C", "C#", "français", "en", "UNKNOWN", "C#", "python", "portuguese"]

def normaliser_sans_librairie(texte):
    # 1. Nettoyage strict
    txt = str(texte).strip().lower()
    
    # 2. Élimination des langages informatiques et valeurs inconnues
    intrus = ["python", "c#", "c", "unknown"]
    if txt in intrus or len(txt) < 2:
        return None
        
    # 3. Si c'est déjà un code ISO valide à 2 lettres (ex: "es", "en")
    if len(txt) == 2:
        return txt
        
    # 4. Pour les mots longs ("espanol", "français", "spanish", "portuguese") :
    # On extrait les 2 premières lettres qui correspondent TOUJOURS au code ISO 639-1
    # de ces langues (es, fr, sp -> es, po -> pt via correction automatique)
    code_graine = txt[:2]
    
    # Correction automatique pour les racines spécifiques (ex: "spanish" -> "es", "portuguese" -> "pt")
    if code_graine == "sp": 
        return "es"
    if code_graine == "po": 
        return "pt"
        
    return code_graine

# Application et nettoyage des doublons
resultat_clean = [normaliser_sans_librairie(item) for item in donnees_brutes]
resultat_final = list(set(code for code in resultat_clean if code is not None))

print(resultat_final)
# Sortie garantie sur n'importe quel ordinateur : ['fr', 'pt', 'es', 'en']
```

```python
def normaliser_liste_langues(liste_brute):
    def normaliser_mot(texte):
        # 1. Nettoyage strict
        txt = str(texte).strip().lower()
        
        # 2. Élimination des langages informatiques et valeurs inconnues
        intrus = ["python", "c#", "c", "unknown"]
        if txt in intrus or len(txt) < 2:
            return None
            
        # 3. Si c'est déjà un code ISO valide à 2 lettres (ex: "es", "en")
        if len(txt) == 2:
            return txt
            
        # 4. Extraction des 2 premières lettres pour les mots longs
        code_graine = txt[:2]
        
        # Correction automatique pour les racines spécifiques
        if code_graine == "sp": 
            return "es"
        if code_graine == "po": 
            return "pt"
            
        return code_graine

    # Application de la logique sur chaque élément de la liste
    resultat_clean = [normaliser_mot(item) for item in liste_brute]
    
    # Suppression des valeurs None et des doublons, puis conversion en liste
    return list(set(code for code in resultat_clean if code is not None))


# --- EXEMPLE D'UTILISATION ---
donnees_brutes = ["espanol", "es", "spanish", "python", "C", "C#", "français", "en", "UNKNOWN", "C#", "python", "portuguese"]

# Tu donnes une liste, il en sort une liste
ma_liste_propre = normaliser_liste_langues(donnees_brutes)

print(ma_liste_propre)
# Sortie : ['fr', 'pt', 'es', 'en']
```
